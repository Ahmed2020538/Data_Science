# SimPle CalCu APP 

```python
print("Wellcome To Simple Calculator:")
print("Please Enter Operation one of   + , - , * , / , //  ::")
print("Oper = Enter The Operation ::")
print("Num1 = Enter The First Number :")
print("Num2= = Enter he Second Number ::")
print("Please Enter E To Closing The CalCulator")
print("============================================")


while True :
    Oper = input ("Enter The Operation ::")
    if Oper == "E" :
        print("The Calculator Is Closinfg::")
        print("ThnX To Use Our CalCu App ::")
        break
    Num1 = input ("Enter The First Number ::")
    Num2 = input("Enter The Second Number ::")


    try :
        
        Num1=int(Num1)
        Num2=int(Num2)
        
        if Oper == "+" :
            print(Num1 , "+" , Num2 , "=" , Num1+Num2)
        elif Oper == "-" :
            print(Num1 , "-" , Num2 , "=" , Num1-Num2)
        elif Oper == "*" :
            print(Num1 , "*" , Num2 , "=" , Num1*Num2)
        elif Oper == "/" :
            if Num2 ==0 : Num2=1
            print(Num1 , "/" , Num2 , "=" , Num1/Num2)
        elif Oper == "//" :
            if Num2==0 : Num2 = 1
            print(Num1 , "//" , Num2 , "=" , Num1//Num2)
        else :
            print(" Error Please Enter One Of Operation ::")
            
            
    except ValueError :
        print("Please Enter A Valid Numbers:")
    print("==============================================================")
```
