```python
# present Cusom data from json file
users = json_res["data"]
user
```

```python
# create CSV_file
with open("ApI.csv"  , "w") as f :
    # definetion File Columns
    writer = csv.DictWriter( f , fieldnames= ["Id" ,"Name" ,"E_Mail" , "Gender" , "Statue" ,"created_at" ,"updated_at"])
    writer.writeheader()
    for user in users :
        Id = user["id"]
        Name = user["name"]
        E_mail = user["email"]
        Gender = user["gender"]
        Statue = user["status"]
        Creation = user["created_at"]
        Update = user["updated_at"]
        writer.writerow({"Id" : Id ,"Name" : Name ,"E_Mail" : E_mail , "Gender" : Gender , "Statue" : Statue ,"created_at" : Creation ,"updated_at" : Update })
        
```
