# 5 :: Create function to create csv_file to save data in it

```python
def get_users_data(url):
    with open("users_data_set.csv" , "w") as f :
        writer = csv.DictWriter( f , fieldnames= [ "ID" , "Name" , "E_mail" , "Gender" , "Status" , "creation" , "Updation"])
        writer.writeheader()
        res = R.get(url)
        res_json = json.loads(res.text)
        users = res_json["data"]
        for user in users :
            U_Id     =user["id"]
            U_Name   =user["name"]
            U_e_mail =user["email"]
            U_Gender =user["gender"]
            U_Status = user["status"]
            U_created_at =user["created_at"]
            U_updated_at =user["updated_at"]
            writer.writerow({"ID":U_Id ,"Name" : U_Name ,"E_mail" :U_e_mail ,"Gender" :U_Gender , "Status": U_Status, "creation":U_created_at ,"Updation" :U_updated_at})
            print(f"U_Id :: {U_Id}\nU_Name :: {U_Name}\nU_E_mail :: {U_e_mail}\nU_Gender :: {U_Gender}\nU_Status :: {U_Status}\nU_Created :: {U_created_at}\nU_Updated :: {U_updated_at}\n-------------------")
            
    
```

```python
get_users_data("https://gorest.co.in/public-api/users?_format=json&access-token=H4lYmIGLXZkDlpy_w0ZVS5rgcpDXbN7fhRRy")
```
