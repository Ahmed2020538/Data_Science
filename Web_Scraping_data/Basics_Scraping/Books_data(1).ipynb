```python
import requests as r
import csv
from bs4 import BeautifulSoup as B
```

```python
def Books_data_set(url) :
    with open("Books_data_setss.csv" , "w+") as f :
        writer = csv.DictWriter( f , fieldnames=["Book_Name" , "Book_Rate" , "Book_Price" , "Book_description"])
        writer.writeheader()
        res = r.get(url)
        soup = B(res.text , "html.parser")
        Rating = {"One" : 1 , "Two" : 2 , "Three" : 3 , "Four" : 4 , "Five" : 5}
        Articles = soup.find_all("article" , attrs = {"class" : "product_pod"})
        for Article in Articles :
            BooK_name = Article.find("h3").get_text()
            BooK_rate = Article.find("p").get("class")[1]
            BooK_price = float(Article.find("p" , attrs = {"class" : "price_color"}).get_text().split("Â£")[1])
            BooK_describe = Article.find("p" , attrs = {"class" : "instock availability"}).get_text().strip()
            writer.writerow({"Book_Name" : BooK_name , "Book_Rate" : Rating[BooK_rate] , "Book_Price" : BooK_price , "Book_description" : BooK_describe})
            print(f"The Book_Name :: {BooK_name}\nThe Book_Rate :: {Rating[BooK_rate]}\n\
The BooK_Price :: {BooK_price}\nThe Book_Description :: {BooK_describe}\n\n=======\n===========\n\n")
```

```python
Books_data_set("http://books.toscrape.com/catalogue/category/books/mystery_3/index.html")
```
