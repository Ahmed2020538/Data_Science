* Create Function To Scrapping data from website 

```python
def get_data (url):
    res = R.get(url)
    soup= B(res.text , "html.parser")
    Articles = soup.find_all("article" , attrs = {"class" : "product_pod"})
    Rating = {"One" : 1 , "Two" :2 ,"Three" : 3 , "Four" : 4 , "Five" : 5}
    for Article in Articles :
        Book_Name = Article.find("h3").get_text()
        Book_Rate = Article.find("p").get("class")[1]
        Book_Price = float(Article.find("p" , attrs ={"class" : "price_color"}).get_text().split("£")[1])
        Book_Aval = Article.find("p" , attrs = {"class" : "instock availability"}).get_text().strip()
        print("====================================================")
        print("====================================================")
        print(f"The Bo_Name :: {Book_Name}\nThe Bo_Rate :: {Rating[Book_Rate]}\nThe Bo_Price :: {int(Book_Price)}\nThe Bo_Aval :: {Book_Aval}")


    
```

```python
get_data("http://books.toscrape.com/catalogue/category/books/mystery_3/index.html")
```

* Create Function To Scrapping data from website 

```python
def get_data2(url):
    res = R.get(url)
    soup = B(res.text , "html.parser")
    Days = soup.find_all("div" , attrs = {"class" : "day"})
    for d in Days :
        day = d.find("div" , attrs = {"class" : "title"}).find("b").get_text()
        date = d.find("div" , attrs = {"class" : "title"}).find("span").get_text()
        Temp_max = d.find("div" , attrs = {"class" : "temps"}).find("b")
        for T in Temp_max :
            Max = T.split(":")[1]
        Temp_min = d.find("div" , attrs = {"class" : "temps"}).find("span").get_text().split(":")[1]
        Pred = d.find("div" , attrs = {"class" : "info"}).find("span").get_text()
    
        Wind =d.find("div" , attrs = {"class" : "wind"}).get_text().split("°")[1]
        wind_popinfo = d.find("div" , attrs = {"class" : "wind-popinfo"}).get_text()
    
        print(f"Day :: {day}\nDate :: {date}\nMin :: {Temp_min}\nMax :: {Max} \nPred :: {Pred}\nwind_popinfo :: {wind_popinfo}\nWind :: {Wind}")
        print("====================================")
        print("====================================")
    
    
```

```python
get_data2("https://eg.freemeteo.com/weather/cairo/7-days/list/?gid=360630&language=english&country=egypt")
```
