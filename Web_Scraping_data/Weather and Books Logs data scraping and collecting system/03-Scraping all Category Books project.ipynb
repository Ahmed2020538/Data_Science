# Scraping data defination ::
*********************************
* Web scraping is a term for various methods used to collect information from across the Internet. Generally, this is done with software that simulates human Web surfing to collect specified bits of information from different websites. Those who use web scraping programs may be looking to collect certain data to sell to other users, or to to use for promotional purposes on a website.

* Web scraping is also called Web data extraction, screen scraping or Web harvesting

* Import Required Libraries

```python
# Import liberaries 
import requests as r
from bs4 import BeautifulSoup as B
```

* Create func To get Book-Names

```python
def Books_name(url) :
    res= r.get(url)
    soup = B(res.text , "html.parser")
    Articles = soup.find_all("article" , attrs = {"class" : "product_pod"})
    for A in Articles :
        Book_Name = A.find("h3").get_text()
        #Book_Names = A.find("h3").find("a").get("title")
        print(f"The Book Name :: {Book_Name}\n")
        
```

* Create Func To get Books_data


```python
def Books_data(url) :
    Rating = {"One" : 1 , "Two" : 2 , "Three" : 3 , "Four" : 4 , "Five" : 5}
    res = r.get(url)
    soup = B(res.text , "html.parser")
    Articles = soup.find_all("article" , attrs = {"class" : "product_pod"})
    for A in Articles :
        Book_name  = A.find("h3").get_text()
        Book_Rate  = A.find("p").get("class")[1]
        Book_Price = float(A.find("p" , attrs = {"class" : "price_color"}).get_text().split("Â£")[1])
        Book_Info  = A.find("p" , attrs = {"class" : "instock availability"}).get_text().strip()
        
        print(f"Book_name :: {Book_name}\nBook_Rate ::{Rating[Book_Rate]}\nBook_Price :: {Book_Price}\nBook_Info :: {Book_Info}")
        print("==============================(Book_details)==========================================")
```

* Create function to get categry_name and its_links


```python

def Categurical_data (url) :
    categuries = {}
    res = r.get(url)
    soup = B(res.text , "html.parser")
    links = soup.find("div" , attrs = {"class" : "side_categories"}).find("ul").find("li").find("ul").find_all("a")
    for link in links :
        org_link = link.get("href").replace("../" , "http://books.toscrape.com/catalogue/category/books/")

        if org_link != "index.html" :
            category_name = link.get_text().replace(" " , "").replace("\n" ,"")
            categuries[category_name] = org_link
            
    return categuries
```

* Present all Books categiries data

```python
Categories = Categurical_data("http://books.toscrape.com/catalogue/category/books/travel_2/index.html")
for Category_Name , Category_Link in Categories.items(): 
    print(f"The Category_Name :: {Category_Name}\n\nThe Category_Link :: {Category_Link}")
    print("\n\n=================================(^ Books ^)================================\n\n=========\n")
    print(f"\n {Books_name(Category_Link)}\n==================\n{Books_data(Category_Link)}")
  
```
