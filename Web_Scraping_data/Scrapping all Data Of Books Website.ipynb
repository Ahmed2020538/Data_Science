# Function OF Scraping data

```python
def get_Books_data (url) :
    res    = R.get(url)
    soup   = B( res.text , "html.parser")
    Rating = { "One" : 1 , "Two" : 2 , "Three" : 3 , "Four" : 4 , "Five" : 5 }
    Arteciles = soup.find_all("article" , attrs = { "class" : "product_pod" })
    for Artecile in Arteciles :
        Book_Name = Artecile.find("h3").get_text()
        Book_Rate = Artecile.find("p").get("class")[1]
        Book_Price = float(Artecile.find("p" , attrs = { "class" : "price_color" }).get_text().split("£")[1])
        print(f"The Bo_Name :: {Book_Name}\nThe Bo_Rate :: {Rating[Book_Rate]}\nThe Bo_Price :: {Book_Price}")
        print("==========================( ^  The Book Detalies  ^ )==============================")
    
```

```python
get_Books_data("http://books.toscrape.com/catalogue/category/books/mystery_3/index.html")
```

```python
def get_all_data_categories (link):
    Categories = {}
    res = R.get(link)
    soup = B( res.text , "html.parser" )
    Links = soup.find("div" , attrs = {"class" : "side_categories"}).find("ul").find("li").find("ul").find_all("a")
    for link in Links :
        Org_url = link.get("href").replace("../" , 'http://books.toscrape.com/catalogue/category/books/' )
        if Org_url != 'index.html' :
            Category_name = link.get_text().replace(' ', '').replace('\n', '')
            Categories[Category_name] = Org_url
        
    return Categories
    
```

```python
Categories = get_all_data_categories("http://books.toscrape.com/catalogue/category/books/travel_2/index.html")
for Category_Name , Link in Categories.items() :
    print(f"\nThe category_Name  :: {Category_Name}\n\n\nThe link  :: {Link}\n\n")
    print("========================( ^  Books  ^)===============================")
    print(f"\n  {get_books(Link)}\nThe detailes of Book :: {get_Books_data(Link)}")
    print("===================================================")
```
