```python
import requests as r
import csv
from bs4 import BeautifulSoup as B
import pandas as pd
```

```python
def get_Books_data (website) :
    with open("Books_dataet.csv" , "w") as f :
        writer = csv.DictWriter(f , fieldnames=["TiTle" , "Price" , "Rate" ,"Aval"])
        writer.writeheader()
        Rating = {"One" : 1 , "Two" : 2 ,"Three" : 3 ,"Four" : 4 ,"Five" : 5 }
        res = r.get(website)
        soup = B(res.text , "html.parser")
        Articles = soup.find_all("article" , attrs = {"class" : "product_pod"})
        for Article in Articles :
            Book_Name = Article.find("h3").find("a").get("title")
            Book_Price = Article.find("p" , attrs = {"class" : "price_color"}).get_text().split("Â£")[1]
            Book_Rate  = Article.find("p").get("class")[1]
            Book_Avla = Article.find("p" , attrs = {"class" : "instock availability"}).get_text().strip()
            writer.writerow({"TiTle" : Book_Name , "Price" : Book_Price , "Rate" : Rating[Book_Rate] , "Aval" :Book_Avla})
            print(f"The Book_Name is :: {Book_Name}\nThe Book_Price is :: {Book_Price}\nThe Book Rate is :: {Rating[Book_Rate]}\nThe Book Aval is :: {Book_Avla}\n-----------------")
```

```python
get_Books_data("http://books.toscrape.com/catalogue/category/books/mystery_3/index.html")
```

```python
def All_Books_Categories (link) :
    categories = {}
    res = r.get(link)
    soup = B(res.text , "html.parser")
    links = soup.find("div" , attrs = { "class" : "side_categories"}).find("ul").find("li").find("ul").find_all("a")
    for link in links:
        url = link.get("href").replace("../" , "http://books.toscrape.com/catalogue/category/books/")
        if url != "index.html" :
            categorie_name = link.get_text().replace(" " ,"").replace("\n" ,"")
            categories[categorie_name] = url
    return categories
```

```python
categories = All_Books_Categories("http://books.toscrape.com/catalogue/category/books/travel_2/index.html")
```

```python
for categorie_name , link in categories.items() :
    print(f"The Category_Name is :: {categorie_name}\nThe Category link is :: {link}\n\n\nThe Books of category  :: {categorie_name} (^_^)  ")
    print("==========================================")
    print(f"The Cotegory Books is :: {get_Books_data(link)}\n ")
```
