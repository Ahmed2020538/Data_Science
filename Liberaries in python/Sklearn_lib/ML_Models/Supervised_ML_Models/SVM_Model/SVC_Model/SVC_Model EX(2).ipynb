```python
# Import Liberaries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report , confusion_matrix
```

```python
# loading dataset

path ="G:\\Data Field\\D..S..Course\\data.sc.Asem\\Data\Data\\2.5 SVC\\heart.csv"
data = pd.read_csv(path)
data
```

* Spliting data into input and output data

```python
X = data.iloc[ : , : -1]
y = data.iloc[ : , -1]

X
```

```python
y
```

* Spliting data into Training and testing data

```python
X_train , X_test , y_train , y_test = train_test_split( X , y , test_size = 0.33 , random_state = 44 , shuffle = True)
```

* Normalize X_train and X_Test Only

```python
Normlize_data = StandardScaler()
X_train = Normlize_data.fit_transform(X_train)
X_test =Normlize_data.fit_transform(X_test)
```

* Appling SVC Model

```python


SVC_Model = SVC( kernel="rbf" , C = 0.1 , gamma="auto" ,random_state = 44)
SVC_Model.fit(X_train , y_train)
Model_training_Score = SVC_Model.score( X_train, y_train)
Model_testing_Score = SVC_Model.score( X_test, y_test)
y_pred = SVC_Model.predict(X_test)

Model_Accuracy_Report = classification_report(y_test , y_pred)
C_M = confusion_matrix(y_test , y_pred)

sns.heatmap(C_M , center=True)
plt.show()
```
