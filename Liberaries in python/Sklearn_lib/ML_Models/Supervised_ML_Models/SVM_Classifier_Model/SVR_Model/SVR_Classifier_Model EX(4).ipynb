# SVR_Classifier_Model 

```python
# Import Liberaries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error , mean_squared_error , median_absolute_error ,r2_score
```

```python
# loading dataset

path = "G:\\Data Field\\D..S..Course\\data.sc.Asem\\Data\Data\\2.4 SVR\\satf.csv"
data = pd.read_csv(path)
data
```

* Handling missing data values


```python

Handlling_data = SimpleImputer(missing_values=np.nan , strategy="mean")
data = Handlling_data.fit_transform(data)
data
```

* Normalizing data using Standerzation model


```python

Rescalling_data = StandardScaler()
data = Rescalling_data.fit_transform(data)
data
```

* Spliting data into input ,, output data becouse its supervised ML_Algo

```python
X = data[ : , : -1]
y = data[ : , -1]

print(f"The X_data :: \n{X}\n=======\nThe y_data :: \n{y}\n======\nThe X_data_shap :: {X.shape}\n====\n\
The y_data_shap :: {y.shape}")
```

* Spliting data into Traning data to fit model and testing data to estimate

```python
X_train , X_test , y_train , y_test = train_test_split( X , y , test_size= 0.33 , random_state = 44 , shuffle = True)

print(f"The X_train :: \n{X_train[:10]}\n========\nThe X_test :: \n{X_test[:10]}\n=======\nThe y_train :: \n{y_train[:10]}\
\n======\nThe y_test :: \n{y_test[:10]}\n======\nThe X_train_shap :: {X_train.shape}\n====\nThe X_test_shap ::\
\n{X_test.shape}\n=====\nThe y_train_shap ::{y_train.shape}\n=====\nthe y_test_shap ::{y_test.shape}")
```

* Appling  SVR_Model 

```python
SVR_model = SVR( C = 20 , kernel= "rbf" , epsilon = 0.01 , max_iter=1000)
SVR_model.fit(X_train , y_train)
Model_Training_Score = SVR_model.score(X_train , y_train)
Model_Tesing_Score = SVR_model.score(X_test , y_test)
y_pred = SVR_model.predict(X_test)

MAE = mean_absolute_error(y_test , y_pred , multioutput="uniform_average")
MSE = mean_squared_error(y_test , y_pred , multioutput="uniform_average")
Mid_A_E = median_absolute_error(y_test , y_pred)
r2_score = np.sqrt(MSE)

```

```python
print(f"The Model_Training_Score :: {Model_Training_Score}\n=======\nThe Model_Tesing_Score ::\
{Model_Tesing_Score}\n=======\nThe y_pred :: \n{y_pred[:10]}\nthe y_real :: \n{y_test[:10]}\n=======\n\
The MAE :: {MAE}\n=====\nThe MSE :: {MSE}\n======\nthe Mid_A_E :: {Mid_A_E}\n======\nThe r2 Score ::\
{r2_score}")
```

* Changing the SVR_Model_hyperparameter values

```python
SVR_model = SVR( C = 20 , kernel= "linear" , epsilon = 0.01 , max_iter=1000)
SVR_model.fit(X_train , y_train)
Model_Training_Score = SVR_model.score(X_train , y_train)
Model_Tesing_Score = SVR_model.score(X_test , y_test)
y_pred = SVR_model.predict(X_test)

MAE = mean_absolute_error(y_test , y_pred , multioutput="uniform_average")
MSE = mean_squared_error(y_test , y_pred , multioutput="uniform_average")
Mid_A_E = median_absolute_error(y_test , y_pred)
r2_score = np.sqrt(MSE)
```
