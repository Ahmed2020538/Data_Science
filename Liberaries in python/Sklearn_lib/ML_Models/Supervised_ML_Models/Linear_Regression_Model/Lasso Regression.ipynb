```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error , mean_squared_error , median_absolute_error
```

```python
Bosted_data = load_boston()

X = Bosted_data.data
y = Bosted_data.target

X_train , X_test , y_train , y_test = train_test_split( X , y , test_size = 0.33 , random_state = 44 ,shuffle = True)

Lasso_Regression_Model = Lasso(fit_intercept = True , copy_X = True , alpha = 1.5 , normalize = True)
Lasso_Regression_Model.fit(X_train , y_train)
Model_coef = Lasso_Regression_Model.coef_
Model_intercept = Lasso_Regression_Model.intercept_

y_pred = Lasso_Regression_Model.predict(X_test)

Accuracy_training = Lasso_Regression_Model.score(X_train , y_train)
Accuracy_Testing = Lasso_Regression_Model.score(X_test , y_test)

MAE = mean_absolute_error(y_test , y_pred , multioutput="uniform_average")
MSE = mean_squared_error(y_test , y_pred , multioutput="uniform_average")
Md_A_E = median_absolute_error(y_test , y_pred )
```
