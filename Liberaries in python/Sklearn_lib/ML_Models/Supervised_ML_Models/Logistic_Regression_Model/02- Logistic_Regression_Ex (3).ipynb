```python
# import liberary
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report , confusion_matrix
```

```python
# Loading data
path = "G:\\Data Field\\D..S..Course\\data.sc.Asem\\Data\Data\\2.2 Logistic Regression\\heart.csv"
data= pd.read_csv(path)
data
```

```python
# Spliting data into Input and Output data
X = data.iloc[ : , : -1]
y = data.iloc[ : , -1]
```

```python
X
```

```python
y
```

```python
# Rescalling data using Standetzation model

Standerzation = StandardScaler()
X  = Standerzation.fit_transform(X)
X
```

```python
# Spliting data into Training and Testing data
X_train , X_test , y_train , y_test = train_test_split( X  , y , test_size = 0.33 , random_state =44 , shuffle = True)

print(f"The X_train :: \n{X_train}\n=======\nThe X_test :: \n{X_test}\n=====\nThe y_train :: \n {y_train}\n=======\n\
The y_test :: \n {y_test}\n====\nThe X_train_shap :: {X_train.shape}\n===\nThe X_test_shape :: {X_test.shape}\
\n=====\nThe y_train_shape :: {y_train.shape}\n=====\nThe y_test_shape :: {y_test.shape}")
```

```python
#Applying LogisticRegression Model 

'''
#linear_model.LogisticRegression(penalty='l2’,dual=False,tol=0.0001,C=1.0,fit_intercept=True,intercept_scaling=1,
#                                class_weight=None,random_state=None,solver='warn’,max_iter=100,
#                                multi_class='warn’, verbose=0,warm_start=False, n_jobs=None)
'''


Logistic_Regression_Model = LogisticRegression(penalty="l2" , solver = "sag" ,max_iter=10000, random_state=33 , C =1.0 )

Logistic_Regression_Model.fit(X_train , y_train)
Model_Coef = Logistic_Regression_Model.coef_
Model_intercept = Logistic_Regression_Model.intercept_
Model_Classes = Logistic_Regression_Model.classes_
Model_iteration = Logistic_Regression_Model.n_iter_
Model_Training_Score = Logistic_Regression_Model.score(X_train , y_train)
Model_Testing_Score = Logistic_Regression_Model.score(X_test , y_test)

y_pred = Logistic_Regression_Model.predict(X_test)
probability = Logistic_Regression_Model.predict_proba(X_test)
Confusion_Matrx = confusion_matrix(y_test , y_pred)
Classifer_report = classification_report(y_test , y_pred)
```

```python
# Present data details

print("The Model Coeffient :: \n" , Model_Coef)
print("=====================================")
print("The Model intercept :: " , Model_intercept)
print("=====================================")
print("The Model_Classes :: " , Model_Classes)
print("=====================================")
print("The Model_Training_Score :: " , Model_Training_Score)
print("=====================================")
print("The Model_Testing_Score :: " , Model_Testing_Score)
print("=====================================")
print("The Model_iteration :: " , Model_iteration)
print("=====================================")
print("The y_pred :: \n" , y_pred)
print("=====================================")
print("The probability :: \n" , probability)
print("=====================================")
print("The Confusion_Matrx ::\n " , Confusion_Matrx)
print("=====================================")
print("The Classifer_report :: " , Classifer_report)
```

```python
# present The Visualization of Confusion_matrix


sns.heatmap(Confusion_Matrx , center=True)
plt.show()
```
