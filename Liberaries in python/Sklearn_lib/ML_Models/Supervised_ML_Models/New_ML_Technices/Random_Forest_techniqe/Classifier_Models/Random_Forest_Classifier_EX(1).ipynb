* Importing Lideraries

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix , classification_report , f1_score
```

* Loading datset

```python
X , y = load_breast_cancer(return_X_y=True)
```

* Spliting data into Tranining and testing data

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.33 , random_state = 44 , shuffle = True)
```

* Appling Randem_Forest_Classifier

```python
for i in range(50 , 500 , 50) :
    Random_Forest_Classifier = RandomForestClassifier( criterion='entropy' , max_depth =2 , n_estimators = i , random_state = 44)
    Random_Forest_Classifier.fit(X_train , y_train)
    Model_classes = Random_Forest_Classifier.classes_
    Fetures_importance = Random_Forest_Classifier.feature_importances_
    Model_Traing_Score = Random_Forest_Classifier.score(X_train , y_train)
    Model_Testing_Score = Random_Forest_Classifier.score(X_test , y_test)
    y_pred = Random_Forest_Classifier.predict(X_test)
    prob_classifier = Random_Forest_Classifier.predict_proba(X_test)
    Conf_Matrix = confusion_matrix(y_test , y_pred)
    Classifier_reporting = classification_report(y_test , y_pred)
    print(f"The n_Estimator :: {i}\n====\nThe Model_classes :: {Model_classes}\n======\nThe Model_Traing_Score :: {Model_Traing_Score}\
\n======\nThe Model_Testing_Score :: {Model_Testing_Score}\n======\nThe y_pred :: {y_pred[:30]}\
\n====\nThe y_test :: {y_test[:30]}\n======\nThe prob_classifier :: \n{prob_classifier[:5]}\n====\n\
The Conf_Matrix ::\n{Conf_Matrix}\n=======\nThe Classifier_reporting :: \n{Classifier_reporting}\
\n======\nThe Fetures_importance :: \n{Fetures_importance}\n=====\n")
```
