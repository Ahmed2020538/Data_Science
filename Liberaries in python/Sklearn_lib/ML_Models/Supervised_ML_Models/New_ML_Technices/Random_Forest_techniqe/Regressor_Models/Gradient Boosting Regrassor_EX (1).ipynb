* Importing Liberaries

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error , mean_squared_error , median_absolute_error , r2_score
```

* loading dataset

```python
X , y = load_boston( return_X_y = True)

print(f"The X_data :: \n{X[:10]}\n========\nThe y_data :: \n {y[:10]}\n======\nThe X_data_shap :: {X.shape}\
\n=============\nThe y_data_shap :: {y.shape}")
```

* Spliting data into Training and Testing data

```python
X_train , X_test , y_train , y_test = train_test_split( X , y , test_size = 0.33 , random_state = 44 , shuffle = True)
print(f"The X_train ::\n{X_train[:5]}\n=======\nThe X_test :: \n{X_test[:5]}\n======\nThe y_Train ::\
\n{y_train[:5]}\n======\nThe y_test ::\n{y_test[:5]}\n=====\nThe X_train_shap :: {X_train.shape}\n======\n\
The X_test :: {X_test.shape}\n=======\nThe y_train_shap :: {y_train.shape}\n=======\nThe y_test ::\n\
{y_test.shape}")
```

* Apply Gradient Boosting Regressor

```python
for i in range (50 , 500 , 50) :
    GB_Regressor = GradientBoostingRegressor(n_estimators=i , random_state =44 , learning_rate =1.5  , max_depth =2)
    GB_Regressor.fit(X_train , y_train)
    Model_Training_Score = GB_Regressor.score(X_train , y_train)
    Model_Testng_Score = GB_Regressor.score(X_test , y_test)
    NUm_feature = GB_Regressor.n_features_
    Ferature_importantece = GB_Regressor.feature_importances_
    y_pred = GB_Regressor.predict(X_test)
    
    MAE = mean_absolute_error(y_test , y_pred , multioutput="uniform_average")
    MSE = mean_squared_error(y_test , y_pred , multioutput="uniform_average")
    Mid_A_E = median_absolute_error(y_test , y_pred)
    R2_score = r2_score(y_test , y_pred)
    
    print(f"The N_Estimator ::{i}\n============\nThe Model_Training_Score :: {Model_Training_Score}\n=======\nThe Model_Testng_Score :: \
    {Model_Testng_Score}\n======\nThe NUm_feature :: {NUm_feature}\n======\nThe Ferature_importantece :: \n\
    {Ferature_importantece}\n=======\nThe y_pred :: {y_pred[:10]}\n=======\nThe y_Test :: {y_test[:10]}\
    The MAE :: {MAE}\n======\nThe MSE :: {MSE}\n=====\nThe Mid_A_E :: {Mid_A_E}\n=======\nThe R2_score::\
    {R2_score}")
    
```
