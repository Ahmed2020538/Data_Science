```python
# Import Liberaries

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error , mean_squared_error , median_absolute_error , r2_score
```

```python
# loading data

X , y = load_boston( return_X_y=True)

#present data 

print(f"The X_data :: \n{X[:10]}\n==============\nThe y_data :: \n{y[:10]}\n==============\n\
The X_data_shap :: {X.shape}\n========\nThe y_data_shap :: {y.shape}")
```

```python
# Spliting data into Training and Testing data

X_train , X_test ,y_train , y_test = train_test_split( X , y , test_size = 0.33 , random_state = 44, shuffle = True)

# present data

print(f"The X_train ::\n{X_train[:10]}\n=============\nThe X_test ::\n{X_test[:10]}\n========\n\
The y_train :: {y_train[:10]}\n========\nThe y_test ::  {y_test[:10]}\n==========\n\
The X_Train_shape :: {X_train.shape}\n========\nThe X_test_shap :: {X_test.shape}\n=======\n\
The y_train_shap :: {y_train.shape}\n=======\nThe y_test_shap :: {y_test.shape}")
```

```python
# Appling Decision Tree Regressor

Decision_Tree_reg = DecisionTreeRegressor(max_depth=3 , random_state = 33)
Decision_Tree_reg.fit(X_train , y_train)
Model_Training_Score = Decision_Tree_reg.score(X_train , y_train)
Model_Testing_Score = Decision_Tree_reg.score(X_test , y_test)
y_pred = Decision_Tree_reg.predict(X_test)

# Present The Accuracy of model

MAE = mean_absolute_error(y_test , y_pred , multioutput="uniform_average")
MSE = mean_squared_error(y_test , y_pred , multioutput="uniform_average")
Mid_A_E = median_absolute_error(y_test , y_pred )
R2_score = r2_score(y_test , y_pred , multioutput="uniform_average")

```

```python
# Present Conclusion

print(f"The Model_Training_Score :: {Model_Training_Score}\n==============\n\
The Model_Testing_Score :: {Model_Testing_Score}\n============\nThe y_pred :: {y_pred}\n==========\n\
The MAE :: {MAE}\n=========\nThe MSE :: {MSE}\n=========\nThe Mid_A_E :: {Mid_A_E}\n======\n\
The R2_score :: {R2_score}\ny_test :: {y_test[:10]}\n=======\n y_pred :: {y_pred[:10]}")
```
