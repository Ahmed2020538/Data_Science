```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
%matplotlib inline
```

* Loading dataset

```python
df = pd.read_csv("data.csv")
df
```

```python
X = df.iloc[ : 500 , :].values
```

```python
X
```

```python
X.shape
```

```python
# Visualize Elbow_Method
lis = []
n = 30
for i in range( 1 , n) :
    kM_Model = KMeans(n_clusters=i)
    kM_Model.fit(X)
    lis.append(kM_Model.inertia_)
    
plt.plot(range(1 , n) , lis)
plt.title("Elbow")
plt.xlabel("Num Of Clusters")
plt.ylabel("inertia_value")
plt.show()
```

```python
kM2_Model = KMeans(n_clusters=5 , init="k-means++")
kM2_Model.fit(X)
y_kmeans = kM2_Model.predict(X)
centers = kM2_Model.cluster_centers_
```

```python
# Visualize clusters and its centroides

sns.set(rc = {"figure.figsize" : [20 , 10]} , font_scale=2)
plt.scatter(X[: , 0] , X[:,1] , c = y_kmeans , cmap="viridis")
plt.scatter(centers[: , 0] , centers[: , 1] , s = 100 , c = "red")
```

```python
# Visualize clusters and its centroides
sns.set(rc = {"figure.figsize" : [20 , 10]} , font_scale=2)
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'm' , label = "Cluster1")
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'b' , label = "Cluster2")
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'g' , label = "Cluster3")
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'r' , label = "Cluster4")
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'y' , label = "Cluster5")
plt.scatter(kM2_Model.cluster_centers_[:, 0], kM2_Model.cluster_centers_[:, 1], s = 300, c = 'black')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
