```python
import matplotlib.pyplot as plt  
import pandas as pd  
import seaborn as sns
```

```python
df = pd.read_csv("shopping_data.csv")
df
```

```python
sns.set(rc = {"figure.figsize" : [20 , 10]} , font_scale = 2)

sns.jointplot("Annual Income (k$)" , "Spending Score (1-100)" , data = df , height=20)
```

```python
X = df.iloc[:, 3:5].values  


import scipy.cluster.hierarchy as shc

sns.set(rc = {"figure.figsize" : [60 , 40]} , font_scale = 6.7)
plt.title("Customer Dendograms")  
dend = shc.dendrogram(shc.linkage(X, method='ward')) 


from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')  
y_labels = cluster.fit_predict(X)
y_labels
```

```python
sns.set(rc = {"figure.figsize" : [20 , 10]} , font_scale = 2)
plt.scatter(X[ : , 0] , X[ : , 1] , c =y_labels , cmap = "rocket")
```

```python
sns.set(rc = {"figure.figsize" : [20 , 10]} , font_scale = 2)
plt.scatter(X[y_labels == 0, 0], X[y_labels == 0, 1], s = 10, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_labels == 1, 0], X[y_labels == 1, 1], s = 10, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_labels == 2, 0], X[y_labels == 2, 1], s = 10, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_labels == 3, 0], X[y_labels == 3, 1], s = 10, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_labels == 4, 0], X[y_labels == 4, 1], s = 10, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
