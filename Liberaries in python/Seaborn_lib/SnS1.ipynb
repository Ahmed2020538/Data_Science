```python
import pandas as pd

import seaborn as sns

data = pd.read_excel("F:\\patition E\\Data Rows\\sales1.xls")

print(data.head(20))

print(data.describe())
```

# Run Scatter plot

```python
sns.relplot("Sales" , "Profit" , data = data)
```

# Run scatter plot using hue

```python
sns.relplot("Sales" , "Profit" , data = data , hue = "Order Priority")
```

# Run scatter plot using Style

```python
sns.relplot("Sales" , "Profit" , data = data , hue = "Order Priority" , style = "Order Priority")
```

```python
sns.relplot("Sales" , "Profit" , data = data , hue = "Order Priority" , style = "Ship Mode")
```

# Run scatter plot using size of  limit disribution

```python
sns.relplot("Sales" , "Profit" , data = data , hue = "Order Priority" ,size = "Discount"
,style = "Order Priority")
```

# Run scatter plot using sizes of points 

```python
sns.relplot("Sales" , "Profit" , data = data , hue = "Order Priority" ,size = "Discount"
,sizes = (50,200),style = "Order Priority")
```
