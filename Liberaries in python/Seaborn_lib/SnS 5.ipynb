```python
import pandas as pd

import seaborn as sns

data = pd.read_excel("F:\\patition E\\Data Rows\\sales1.xls")
data
```

```python
data.describe()
```

# To draw box_plot of data

```python
sns.catplot("Product Category","Shipping Cost",data = data , kind = "box")
```

# run box_plot using X _Axes  and Y _Axes

```python
sns.catplot(x = "Product Category",y = "Shipping Cost",data = data , kind = "box" )
```

# run Box_plot using hue parameters ("Product Category" , "Ship Mode")

```python
sns.catplot(x = "Product Category",y = "Shipping Cost",data = data , hue = "Order Priority", kind = "box" )
```

```python
sns.catplot(x = "Product Category",y = "Shipping Cost",data = data , hue = "Ship Mode", kind = "box" )
```

# Order the hue level

```python
sns.catplot(x = "Product Category",y = "Shipping Cost",data = data , hue = "Ship Mode",
hue_order = ["Delivery Truck","Regular Air" , "Express Air"], kind = "box" )
```

# run box_plot  based on Column

```python
sns.catplot(x = "Product Category",y = "Shipping Cost",data = data , col = "Ship Mode", kind = "box" )
```

# run box_plot  based on row

```python
sns.catplot(x = "Product Category",y = "Shipping Cost",data = data , row = "Ship Mode", kind = "box" )
```

# run box_plot  based on Column and row

```python
sns.catplot(x = "Product Category",y = "Shipping Cost",data = data , col = "Ship Mode",row = "Customer Segment", kind = "box" )
```
