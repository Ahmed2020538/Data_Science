```python
import pandas as pd

import seaborn as sns

data = pd.read_excel("F:\\patition E\\Data Rows\\sales1.xls")

data.head()

```

```python
data["year"] = data["Order Date"].dt.year
```

```python
data.year.unique()
```

```python
sns.relplot("year" , "Sales" , data =data , kind = "line")
```

# Ci used to appear or dis the Frame lines

```python
sns.relplot("year" ,"Sales" , data = data , kind = "line" , ci = None)
```

```python
sns.relplot("year" ,"Sales" , data = data , kind = "line" , ci = "sd")
```

# estimator used to determine limitation of lines

```python
sns.relplot("year" ,"Sales" , data = data , kind = "line" , estimator = None)
```

```python
sns.relplot("year" ,"Sales" , data = data , kind = "line" , estimator = "sum")
```

```python
sns.relplot("year" ,"Sales" , data = data , kind = "line" , hue = "Order Priority" ,sizes = (100,200))
```
