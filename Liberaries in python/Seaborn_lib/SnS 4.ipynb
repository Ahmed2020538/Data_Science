```python
import pandas as pd

import seaborn as sns 

data = pd.read_excel("F:\\patition E\\Data Rows\\sales1.xls")

data.head()
```

```python
data.describe()
```

# To draw ordinary scatter plots 

```python
sns.relplot("Sales" , "Profit" , data = data)
```

```python
sns.relplot("Ship Mode" , "Sales" , data = data)
```

# To draw  categorical scatter plots

```python
sns.catplot("Ship Mode" , "Sales" , data = data)
```

# Control jitter

```python
sns.catplot("Ship Mode" , "Sales" , data = data , jitter = False)
```

# To achive Condition

```python
sns.catplot("Ship Mode" , "Sales" , data = data.query("Profit > 1000") , jitter = False)
```

# To draw Swarm plots

```python
sns.catplot("Ship Mode" , "Sales" , data = data.query("Profit > 1000") , kind = "swarm")
```

```python
sns.catplot(  "Sales" ,"Ship Mode", data = data.query("Profit > 1000") , kind = "swarm")
```
