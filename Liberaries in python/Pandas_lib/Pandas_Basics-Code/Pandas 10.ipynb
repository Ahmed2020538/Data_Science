# 2 _ Pilot_Table

```python
import pandas as pd

path = "E:\\Data Rows\\Code B\\Pivot_data_Table.txt"

D_F = pd.read_csv(path)
D_F
```

```python
dff1 =  D_F.pivot_table( index = "date" , columns= "city")
dff1
```

```python
dff2 =  D_F.pivot_table( index = "city" , columns= "date")
dff2
```

# Margains  argument  defualt get mean (avrage

```python
dff3 =  D_F.pivot_table( index = "city" , columns= "date" , margins=True)
dff3
```

# Aggrate argument


```python
dff4 =  D_F.pivot_table( index = "city" , columns= "date" , margins=True , aggfunc="sum")
dff4
```

```python
dff5 =  D_F.pivot_table( index = "city" , columns= "date" , margins=True , aggfunc="mean")
dff5
```

```python
dff6 =  D_F.pivot_table( index = "city" , columns= "date" , margins=True , aggfunc="std")
dff6
```
