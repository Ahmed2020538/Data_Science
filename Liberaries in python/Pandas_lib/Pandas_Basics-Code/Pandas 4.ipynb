# Hadling missing values 

# 1_ Fillna Func

```python
import pandas as pd 

path = "E:\\Data Rows\\Code B\\Missing_data.txt"

D_F = pd.read_csv(path , parse_dates=["day"])

D_F
```

```python
H_D_F = D_F.fillna(0)     # in all data Convert Nan into 0

H_D_F
```

```python
H_D_F = D_F.fillna({"temperature" : 50 , "windspeed" : 15 , "event" : "Good"} )

H_D_F       # in Special columns convert NaN into
```

```python
H_D_F = D_F.fillna(method="ffill")     # in all data Convert Nan into front value  defualt Rows value

H_D_F
```

```python
H_D_F = D_F.fillna(method="ffill" , axis="columns")     # in all data Convert Nan into front value  By colun

H_D_F
```

```python
H_D_F = D_F.fillna(method="bfill")     # in all data Convert Nan into back value defult in rows

H_D_F
```

```python
H_D_F = D_F.fillna(method="bfill" ,axis="columns")     # in all data Convert Nan into back value  by columns

H_D_F
```

```python
H_D_F = D_F.fillna(method="ffill" , limit = 2)     # in all data Convert Nan into frount value by limit

H_D_F
```

```python
H_D_F = D_F.fillna(method="bfill" ,limit=2)     # in all data Convert Nan into back value by limit

H_D_F
```

# 2 _dropna Func

```python
import pandas as pd 

path = "E:\\Data Rows\\Code B\\Missing_data.txt"

D_F = pd.read_csv(path , parse_dates=["day"])

D_F
```

```python
H_D_F = D_F.dropna()  # remove all rows include NaN Value
H_D_F
```

```python
H_D_F = D_F.dropna(how="all")  # remove all rows that is  all values  include NaN Value
H_D_F
```

```python
H_D_F = D_F.dropna(thresh=1)  # remove all rows that is  include 2 Nan and More
H_D_F
```

```python
H_D_F = D_F.dropna(thresh=2)  # remove all rows that is  include 3 Nan and More
H_D_F
```
