```python
import numpy as np
A = np.array(np.arange(24))

print(A)
```

```python
B= A.reshape(3,8)
print(B)
```

```python
C = A.reshape((2,3,4))
print(C)
```

# Flip left and right

```python
print(np.fliplr(C))  # must 2_dimentional Arr
print("============================")
print(np.fliplr(B))   
```

# Flip Up and down

```python
print(np.flipud(C))  # must 2_dimentional Arr
print("============================")
print(np.flipud(B))  # must 2_dimentional Arr
```

# Roll arry

```python
print(np.roll(A,5))   # use at any dimentional +   take pos or negt prameter 
print("=================================================")
print(np.roll(A,-5))   # use at any dimentional -     take pos or negt prameter 
print("=================================================")
print(np.roll(B,2))     # use at any dimentional +    take pos or negt prameter 
print("=================================================")
print(np.roll(B,-2))   # use at any dimentional -    take pos or negt prameter 
print("=================================================")
print(np.roll(C,2))    # use at any dimentional +    take pos or negt prameter 
print("=================================================")
print(np.roll(C,-2))   # use at any dimentional-    take pos or negt prameter 
print("=================================================")
print(np.rollaxis(C,0,2))
print("=================================================")
print(np.rollaxis(C,1,1))
print("=================================================")
print(np.rollaxis(C,2,0))
```

# Transpose Fun arr

```python
print(np.transpose(A))  # 1_dimentional
print("==============================================")
#print(np.transpose( B , axes = (0,2,1)))  # 2_dimentional Error
print("==============================================")
print(np.transpose( C , axes = (0,2,1)))  # 2_dimentional
```

# Swapaxes fun

```python
print(np.swapaxes(B,1,0))  # for column
print("==============================================")
print(np.swapaxes(B,0,1))  # for column
print("==============================================")
print(np.swapaxes(B,0,0))  # for rows
print("==============================================")
print(np.swapaxes(B,1,1))  #for rows
print("==============================================")
print(np.swapaxes(C,1,0))  # for colums
print("==============================================")
print(np.swapaxes(C,1,1))  # for rows
print("==============================================")
print(np.swapaxes(A,0,0))
print("==============================================")
#print(np.swapaxes(A,1,0))  # have 1_dimentional
```

# Till Fun arr

```python
W = np.array(np.arange(12))
print(W)
```

```python
print(np.tile(W,4))
print("===============================================================")
print(np.tile(W,3))
print("===============================================================")
print(np.tile(W,3).reshape(3,12))
print("===============================================================")
print(np.tile(W,4).reshape(4,12))


```

```python
Z = np.array(np.arange(7))
print(Z)
print("===============================================================")
T1=np.tile(Z,(3,1))
print(T1)
print("===============================================================")
T2 = np.tile(T1,(2,2))
print(T2)
print("===============================================================")
T3 = np.tile(T2,(3,3))
print(T3)
```

# REPEAT FUN ARR

```python
print(Z)
```

# 1_dimentional repeat

```python
print(np.repeat(Z,2))
print("===============================================================")
print(np.repeat(Z,3))
```

# 2_dimentional repeat

```python
print(C)
```

```python
print(np.repeat(C,2,axis=0))  # axis = 0  for rows and Columns repeat    and 2 num of repeat
print("===============================================================")
print(np.repeat(C,3,axis=0))  # axis = 0  for rows and Columns repeat    and 3 num of repeat
```

```python
print(np.repeat(C,2,axis=1))  # axis =1 for Rows only   and  2  num of repeat
print("===============================================================")
print(np.repeat(C,3,axis=1))   # axis =1 for Rows only   and  3  num of repeat
```

```python
print(np.repeat(C,2,axis=2))  # axis =2 for Columns only   and  2  num of repeat
print("===============================================================")
print(np.repeat(C,3,axis=2))   # axis =2 for Columns only   and  3  num of repeat
```
