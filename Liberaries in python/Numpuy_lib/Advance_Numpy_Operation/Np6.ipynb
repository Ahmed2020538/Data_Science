# Linear algebra 

```python
# X + 2*Y = 8
# 3*X + 4*Y = 18

# Solving linear algebra equation

import numpy as np
A = np.array([[1,2] , [3,4]])   # cofficientMatrix
B = np.array([8,18])   # Constants Matrix

print(np.linalg.solve(A,B))

```

```python
import numpy as np

A = np.array([[6,1,1],
              [4,-2,5],
              [2,8,7]])


print(A)
print("========================")

# Ranke of matrix 

print(np.linalg.matrix_rank(A))
print("========================")

# Determinant of matrix
print(np.linalg.det(A))
print("========================")

# inverse of matrix
print(np.linalg.inv(A))
print("========================")


# Exponeniation of matrix
print(np.linalg.matrix_power(A,3))
print("========================")

# To trace matrix

print(np.trace(A))
```

```python
import numpy as np
A_M = np.matrix([[ 3 , 1 , 4 ],[ 1 , 5 , 9 ],[ 2 , 6 , 5 ]])
print(A_M)
print("===========================")
print(A_M.T)
print("===========================")
print(A_M.I)
print("===========================")
print(A_M.H)
print("===========================")
print( A_M * A_M.T)
print("===========================")
print( A_M * A_M.I)
print("===========================")
print( A_M * A_M.H)

print("===========================")
print("===========================")
print("===========================")
A_M_R = np.matrix([[11,22,33],
                   [33,22,11],
                   [22,33,22]])
print(A_M_R)
print("===========================")
A_M_Inv = A_M.I
N_M = A_M_R * A_M_Inv
print(N_M)

print("===========================")
print("===========================")
print(N_M[0])
print("===========================")
print(N_M[1])
print("===========================")
print(N_M[2])
print("===========================")
print("===========================")
print( A_M * N_M - A_M.I)
print("===========================")
print("===========================")
print(N_M * A_M + A_M.H)

```

# Solve Func

```python
from numpy.linalg import solve
print(solve( A_M , A_M_R))
print("===========================")
print("===========================")
print(solve( A_M , A_M_Inv))
print("===========================")
print("===========================")
print(solve( A_M_R , A_M_Inv))
```
