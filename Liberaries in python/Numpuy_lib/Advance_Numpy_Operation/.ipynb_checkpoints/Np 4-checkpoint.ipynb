# Element - Wise - operation

```python
import numpy as np

A = np.array([ [ 1 , 2 , 3 , 4 ],
               [ 5 , 6 , 7 , 8 ] ])
print(A)
print("=================================")
print(A +1)
print("=================================")
print(A**2)
print("=================================")

print(A+2)
print("=================================")

print(A-1)
```

# UNary - Operation

```python
import numpy as np

A = np.array([ [ 1 , 2 , 3 , 4 ],
               [ 5 , 6 , 7 , 8 ] ])

print(A.sum())
print("=================================")


print(A.sum(axis=0))
print("=================================")

print(A.sum(axis=1))
print("=================================")

print(A.max(axis=0))
print("=================================")

print(A.max(axis=1))
print("=================================")

print(A.min(axis=0))
print("=================================")

print(A.min(axis=1))
print("=================================")

print(A.max())
print("=================================")

print(A.min())
```

# Universal Function (Ufunc)

```python
import numpy as np
A=np.array([[1,2,3],
            [4,5,6]])
B=np.array([[4,5,6],
            [1,2,3]])

print(A)
print("=================================")
print(B)
print("=================================")
print(np.sin(A))
print("=================================")
print(np.cos(A))
print("=================================")
print(np.exp(A))


print("=================================")

print(A)
print("=================================")
print(B)
print("=================================")
print(np.sin(B))
print("=================================")
print(np.cos(B))
print("=================================")
print(np.exp(B))
```

# Binary Operation

```python
import numpy as np
A=np.array([[1,2,3],
            [4,5,6]])
B=np.array([[4,5,6],
            [1,2,3]])
print(A)
print("=================================")
print(B)
print("=================================")
print(A+B)
print("=================================")
print(A-B)
print("=================================")
print(A*B)
print("=================================")
print(B-A)
```
