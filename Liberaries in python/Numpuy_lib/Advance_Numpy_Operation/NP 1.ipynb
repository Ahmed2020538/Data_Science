# Array CreaTion 1

```python
# Its array cosist of 2 >>>>>rows  and 1 >> colums

import numpy as np
A1 = np.array([2,6])
print(A1)
print(A1.shape)
print("=================")

# Its arry consist 1 >> row and 2 >> olumns

A2 = np.array([[2,6]])
print(A2)
print(A2.shape)
print("=================")

 # Its array consist of 3 >>rows ana 3 >>columns 

A3 = np.array([[2,6],[1,3],[4,8]])
print(A3)
print(A3.shape)
print("========================")




# Its array consist of 5 >> rows and 1 >> columns

A4 = np.array([2,4,3,5,8])
print(A4)
print(A4.shape)


# Its array consist of 1 >> rows and 5 >> columns


A4 = np.array([[2,4,3,5,8]])
print(A4)
print(A4.shape)
```

# Array CreaTion 2


```python
# create Array by Ay number and  size

#

A = np.zeros(3)
print(A)
print(A.shape)
print("=================")

B = np.zeros((3,3))
print(B)
print("===================")

C = np.ones(5)
print(C)
print(C.shape)
print("===================")

D = np.ones((4,4))
print(D)
print(D.shape)
print("====================")

E = np.ones((3,3),dtype="complex")
print(E)
print("=======================")

F = np.zeros((3,3),dtype="complex")
print(F)
print("===================")

# create arry consist any number by any shape

A1 = np.full((3,3) ,5)

print("A1" ,A1)
print("=================")

# To create Array by range 
# range take int only  But arange Take int or float

A2 = np.arange(10)
A3 = np.arange(1,10,2)

print("A2" ,A2)
print(A2.shape)

print("=================")

print("A3",A3)
print(A3.shape)

print("=================")

# To Create Array by random numbers 

A4 = np.random.random((2,2) )

print("A4",A4)
print("=================")






# To create Array by random int Numbers

A5 = np.random.randint(1,10,(3,3))
A6 = np.random.randint(0,10,(2,2))

print("A5", A5)
print("=================")
print("A6",A6)
print("=================")

# To Create Array using Linespace
A7 = np.linspace(1,10,20)
print("A7",A7)
```
