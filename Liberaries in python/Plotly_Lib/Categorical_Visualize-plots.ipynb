```python
px.box(df1 ,  x  = "species" ,
       y  = "sepal_length" ,
       title = "Box Plot of species & sepal_length" , color_discrete_sequence = px.colors.sequential.Bluyl_r )
```

```python
px.box(df2 ,  x  = "smoker" ,
       y  = "total_bill" ,
       color = "day",
       title = "Box Plot of smoker & total_bill" , 
       color_discrete_sequence = px.colors.sequential.BuGn_r)
```

```python
px.violin(df2 ,  x  = "smoker" ,
       y  = "total_bill" ,
       color = "day",
       title = "Box Plot of smoker & total_bill" , 
       color_discrete_sequence = px.colors.sequential.BuPu_r)
```

```python
px.histogram(df2 ,  x  = "smoker" ,
       y  = "total_bill" ,
       color = "day",
       title = "Box Plot of smoker & total_bill" , 
        barmode="group" ,
       color_discrete_sequence = px.colors.sequential.Burg_r)
```
