```python
import plotly.graph_objects as go
stock_data = px.data.stocks()
stock_data
```

```python
px.line(stock_data , x = "date" , y = "GOOG" , labels={"x" : "Date" , "y"  : "Price"} , title="Time series ")
```

```python
px.line(stock_data , x = "date" , y = ["GOOG" , "AAPL"] , labels={"x" : "Date" , "y"  : "Price"} , title="Time series ")
```

```python
px.line(stock_data , x = "date" , y = ["GOOG" , "AAPL","NFLX" , "FB"] , labels={"x" : "Date" , "y"  : "Price"} , title="Time series ")
```

```python
import plotly.graph_objs as go
```

```python
fig = go.Figure()


fig.add_trace(go.Scatter(x = stock_data.date , 
                         y = stock_data.AMZN , 
                         mode = "lines+markers" , 
                         name = "Amazon"))
fig.add_trace(go.Scatter(x = stock_data.date , 
                         y = stock_data.GOOG , 
                         mode = "lines" , 
                         name = "Google" , 
                        line = dict(color = "seagreen" , width = 2 , dash = "dashdot")))
```

```python
fig.add_trace(go.Scatter(x = stock_data.date , 
                         y = stock_data.AMZN , 
                         mode = "lines+markers" , 
                         name = "Amazon"))
```
