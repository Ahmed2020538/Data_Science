# To draw pie chart plot

```python
import matplotlib.pyplot as plt

exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

plt.pie( exp_vals , labels = exp_vals)
```

```python
plt.pie(exp_vals , labels = exp_vals , shadow = True)
plt.axis("equal") # to icrease the dgree of color chart
plt.show()
```

# to represent the presentage Values

```python
plt.pie(exp_vals , labels = exp_vals , shadow = True , radius =1.5  , autopct ='%1.1f%%' )
plt.axis("equal") # to icrease the dgree of color chart
plt.show()
```

# to explode pie chart 

```python
plt.pie(exp_vals , labels = exp_vals ,shadow = True , radius =1.5, autopct ='%1.1f%%',explode=[0,0,0,0.1,0.2])
plt.axis("equal") # to icrease the dgree of color chart
plt.show()
```

```python
plt.pie(exp_vals , labels = exp_vals ,shadow = True , radius =1.5, autopct ='%1.1f%%',explode=[0,0,0,0.1,0.2],
counterclock=True)
plt.axis("equal") # to icrease the dgree of color chart
plt.show()
```

```python
plt.pie(exp_vals , labels = exp_vals ,shadow = True , radius =1.5, autopct ='%1.1f%%',explode=[0,0,0,0.1,0.2],
counterclock=True, startangle=45)
plt.axis("equal") # to icrease the dgree of color chart
plt.show()
```
