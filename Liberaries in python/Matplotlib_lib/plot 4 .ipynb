```python
import matplotlib.pyplot as plt

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

plt.hist(blood_sugar , bins = 4 , histtype= "step", color = "g")
```

```python
import matplotlib.pyplot as plt

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.xlabel("blood_suger")
plt.ylabel("patient")
plt.title("Blood suger chart")

plt.hist(blood_sugar , bins = 4 , rwidth = 0.75 , histtype = "stepfilled", color = "r" , label ="men_suger")
plt.legend()
```

# to draw multi histopgram in same plot

```python
import matplotlib.pyplot as plt

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]


plt.xlabel("blood_suger")
plt.ylabel("patient")
plt.title("Blood suger chart")

plt.hist([blood_sugar ,blood_sugar_women ] ,
bins =4, rwidth = 0.75 , color = ["r","g"] , label =["men_suger","blood_sugar_women"])
plt.legend()
```
