# marker to point of intercept   and  markersize to size of it  

```python
import matplotlib.pyplot as plt

%matplotlib inline

x=[1,2,3,4,5,6,7]


y=[50,51,52,48,47,49,46]

(plt.plot(x,y, color = "red" , linewidth=5 ,marker = "v", linestyle= "--" ))
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

x=[1,2,3,4,5,6,7]


y=[50,51,52,48,47,49,46]

(plt.plot(x,y, color = "red" , linewidth=2 ,marker = "+" , markersize = 20, linestyle= "--" ))
```

# 4 the degree of line color

```python
import matplotlib.pyplot as plt

%matplotlib inline

x=[1,2,3,4,5,6,7]


y=[50,51,52,48,47,49,46]

(plt.plot(x,y, color = "red" , linewidth=5 ,marker = "v", linestyle= "--" , alpha = 1 ))
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

x=[1,2,3,4,5,6,7]


y=[50,51,52,48,47,49,46]

(plt.plot(x,y, color = "red" , linewidth=5 ,marker = "v", linestyle= "--", alpha = 0.55 ))
```
