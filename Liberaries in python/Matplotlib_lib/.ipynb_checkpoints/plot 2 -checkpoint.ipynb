```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]


plt.xlabel("dayes")
plt.ylabel("tempreture")
plt.title("weather")

x = plt.plot(days ,max_t , label = "max_t" )
y = plt.plot(days, min_t , label = "min_t")
z = plt.plot(days ,avg_t , label = "avg_t" )

print(z,y,z)
```

# THE LEGEND

```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]


plt.xlabel("dayes")
plt.ylabel("tempreture")
plt.title("weather")

x = plt.plot(days ,max_t , label = "max_t" )
y = plt.plot(days, min_t , label = "min_t")
z = plt.plot(days ,avg_t , label = "avg_t" )

plt.legend(loc = "best" , shadow = True  , fontsize = "large")

print(z,y,z)

```

# the grid of  plot

```python
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]


plt.xlabel("dayes")
plt.ylabel("tempreture")
plt.title("weather")

x = plt.plot(days ,max_t , label = "max_t" )
y = plt.plot(days, min_t , label = "min_t")
z = plt.plot(days ,avg_t , label = "avg_t" )

plt.legend(loc = "best" , shadow = True  , fontsize = "large")

plt.grid()
print(z,y,z)

```
