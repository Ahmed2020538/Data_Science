# to draw Bar chart

```python
import matplotlib.pyplot as plt

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]

plt.xlabel("company")
plt.ylabel("revenue")
plt.title("company behavior")

plt.bar(company , revenue , label = "revenue")
plt.legend(loc = "best", fontsize = "small")
```

# to draw muli bar char in the same plot

```python
import matplotlib.pyplot as plt

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]


plt.xlabel("company")
plt.ylabel("revenue")
plt.title("company behavior")

plt.bar(company , revenue , label = "revenue")

plt.bar(company , profit , label = "profit")
plt.legend(loc = "best", fontsize = "small")
```

```python
import matplotlib.pyplot as plt
import numpy as np


company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
z = np.arange(len(company))

plt.xlabel("company")
plt.ylabel("revenue")
plt.title("company behavior")

plt.bar(z-0.2 , revenue , label = "revenue")

plt.bar(z+0.2 , profit , label = "profit")
plt.legend(loc = "best", fontsize = "small")
```

```python
import matplotlib.pyplot as plt
import numpy as np


company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
z = np.arange(len(company))

plt.xlabel("company")
plt.ylabel("revenue")
plt.title("company behavior")

plt.bar(z-0.2 , revenue ,width = 0.4, label = "revenue")

plt.bar(z+0.2 , profit ,width = 0.4 , label = "profit")
plt.legend(loc = "best", fontsize = "small")
```
