```python
# From Date Feature we can generate years , months , days , hours and mintues

df["Years"]   = df["Date"].dt.year
df["Months"]  = df["Date"].dt.month
df["Days"]    = df["Date"].dt.day
df["Weeks"]   = df["Date"].dt.week
df["Months_name"] = df["Date"].dt.month_name()
df["week_ends"]   = df["Date"].dt.weekday
df["Day_names"]   = df["Date"].dt.day_name()
```

```python
df["Postcode"] = df["Postcode"].apply( lambda x  : int(x))
df[["Postcode"]]
```

```python
def Post_type(x) :
    x =str(x)
    if x.endswith(("1" or "2" or "3") ):
        return "Class_A"
    elif x.endswith(("4" or "5" or"6")) :
        return "Class_B"
    elif x.endswith(("7" or "8" or "9" )) :
        return "Class_C"
                  
                   
```

```python
df["Post_Class"] = df["Postcode"].apply(Post_type)
```

* From the result above we found that ::
     * car_Col is categorical so that we will drop na values becouse they are less num
     * CouncilArea and Post_Class col include hue numbers of missing value so that on that case we will fill Becouse if we drop na_values we will Lose lage num of data
     * YearBuilt and BuildingArea Numerical columns and include hue numbers of missing data so that we will fiil it using median Becouse may we found Outliers data so we cont using mean imputer
        

```python
# Solve Car_col_missing data by droping na values
df.dropna(subset=["Car"] , axis = 0 , inplace=True)
# Solve CouncilArea _ cols whiche include hue numbers of null values by filling with mod_value which categorical_data
df['Post_Class'].fillna(df['Post_Class'].mode()[0], inplace=True)
df['CouncilArea'].fillna(df['CouncilArea'].mode()[0], inplace=True)
# Solve YearBuilt and BuildingArea which include hue numbers of null values
Numerical_cols = ["BuildingArea" , "YearBuilt"] 
for col in Numerical_cols :
    df[col].fillna(df[col].median() , inplace=True)
```
