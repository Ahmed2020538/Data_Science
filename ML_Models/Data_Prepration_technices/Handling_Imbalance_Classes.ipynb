# Under_Sample_Technice

```python
# Solving That Problem By doing Undersampling

from sklearn.model_selection import train_test_split
x = df.iloc[ : , : -1]
y = df.Class

x_train , x_test , y_train , y_test = train_test_split( x , y , test_size = 0.25 , random_state = 27)

Train_data = pd.concat([x_train , y_train] , axis=1)
Train_data
```

```python
from sklearn.utils import resample
fruad = Train_data[Train_data["Class"] == 1]
no_fruad = Train_data[Train_data["Class"] == 0]

# Applying UnderSampling
Not_Fruad = resample(no_fruad , replace = False , n_samples = len(fruad))
under_sample = pd.concat([Not_Fruad ,fruad ])
under_sample
```

# Over_Samplling_Technice

```python
under_sample.Class.value_counts()
```

```python
# Appling Oversampling technice
from sklearn.utils import resample
fruad = Train_data[Train_data["Class"] == 1]
no_fruad = Train_data[Train_data["Class"] == 0]

# Applying UnderSampling
Fruad = resample(fruad , replace = True , n_samples = len(no_fruad))
under_sample1 = pd.concat([no_fruad ,Fruad ])
under_sample1
```
