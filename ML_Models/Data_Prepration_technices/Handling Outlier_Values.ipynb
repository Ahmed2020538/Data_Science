```python
df = pd.read_csv("boston_house_prices.csv")
df
```

```python
# To Detect Outliers We will Visualize data using boxplot and strip_plot ::
sns.boxplot( x = "DIS" , data = df)
sns.stripplot( x = "DIS" , data = df  , color="r")
```

```python
# Handlling outliers using replace that values by Median_values
from datasist.structdata import detect_outliers
for col in df.columns :
    Outliers = detect_outliers(df , 0 , [col])
    median_col_value = df[col].median()
    df[col].iloc[Outliers]  = median_col_value
```

```python
# To Detect Outliers We will Visualize data using boxplot and strip_plot ::
sns.boxplot( x = "DIS" , data = df)
sns.stripplot( x = "DIS" , data = df  , color="r")
```
