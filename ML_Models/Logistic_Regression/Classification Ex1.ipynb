```python
def cost_func(thetav , Xv , yv) :
    Xv = np.matrix(Xv)
    yv = np.matrix(yv)
    thetav = np.matrix(thetav)
    first = np.multiply( -yv , np.log(sigmoid(Xv * thetav.T)))
    second = np.multiply((1-yv) , np.log((1-sigmoid(Xv * thetav.T))))
    return np.sum(first - second) / (len(Xv))
```

```python
cost = cost_func(theta , X , y)
```

```python
cost
```

```python
def gradiant_func (thetav , Xv , yv) :
    thetav = np.matrix(thetav)
    Xv = np.matrix(Xv)
    yv = np.matrix(yv)
    
    parameter = int(thetav.ravel().shape[1])
    grad = np.zeros(parameter)
    error = sigmoid(Xv * thetav.T)-yv
    for i in range(parameter) :
        termp = np.multiply(error , Xv[: ,i])
        grad[i] = np.sum((termp) / len(Xv))
        print(f"\n the i = {i}\n\n grad[i] :: {grad[i]}")
    return grad
```

```python
import scipy.optimize as opt
res = opt.fmin_tnc(func = cost_func  , x0 = theta , fprime=gradiant_func , args=(X , y))
```

```python
res
```

```python
new_jcost = cost_func(res[0] , X , y)
```

```python
new_jcost
```

```python
def prediction (theta , X) :
    prob = sigmoid(X * theta.T)
    return [ 1 if x >= 0.5 else 0 for x in prob]
```

```python
theta_min = np.matrix(res[0])
predict = prediction(theta_min , X)
```

```python
predict
```

```python
corrct =[ 1  if ((a==0 and b ==0) or (a==1 and b== 1)) else 0 for (a , b) in zip(predict , y)]
```

```python
accuracy = (sum(map(int , corrct)) % len(corrct))
print(f"the Acuuracy = {accuracy}%")
```
