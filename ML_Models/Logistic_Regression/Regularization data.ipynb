```python
lr = 2
```

```python
def sig(z) :
    return 1 / (1+np.exp(-z))
```

```python
def cost(theta , X , y , lr) :
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    first = np.multiply(-y , np.log(sig(X * theta.T)))
    second = np.multiply((1-y) , np.log(1 - sig(X * theta.T)))
    reg = (lr / 2*len(X)) * np.sum(np.power(theta[ : , 1 : theta.shape[1]] , 2))
    return np.sum(first - second) / (len(X) )+ reg
```

```python
cost(theta , X , y , lr)
```

```python
def gradiant(theta , X , y , lr) :
    theta = np.matrix(theta)
    X = np.matrix(X)
    y = np.matrix(y)
    parameter = int(theta.ravel().shape[1])
    grad = np.zeros(parameter)
    error = sig(X*theta.T)
    for i in range(parameter):
        
        term = np.multiply(error , X[ : , i])
        if i==0 :
            grad[i] = np.sum(term) / len(X)
        else :
            grad[i] = (np.sum(term) / len(X)) +np.sum((lr /len(X)) * theta[ : , i])
        
    return grad
```

```python
res = opt.fmin_tnc(func=cost, fprime=gradiant , x0 = theta , args=(X , y , lr))
```

```python
def predictive( theta , X) :
    prop = sig(X * theta.T)
    return [ 1 if i >= 0.5 else 0 for i in prop]
min_theta = np.matrix(res[0])
prediction = predictive(min_theta , X)
```
