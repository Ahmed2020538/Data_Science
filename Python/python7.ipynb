```python
import statistics as st
print("The Mean is ::",st.mean([1 ,2 ,3 , 4 , 5]))
print("======================")
print("The Median is ::",st.median([1 , 2 , 3 , 4 , 5]))
print("======================")
print("The Median_grouped is::",st.median_grouped([1,2,3,4,5]))
print("======================")
print("The Median_high is ::",st.median_high([1,2,3,4,5]))
print("======================")
print("The Median low is ::",st.median_low([1,2,3,4,5]))
print("======================")
print("The Pstdev is ::",st.pstdev([1,2,3,4,5]))
print("======================")
print("The Pvariance is ::",st.pvariance([1,2,3,4,5]))
print("======================")
print("The Stdv is ::",st.stdev([1,2,3,4,5]))
print("======================")
print("The Variance is ::",st.variance([1,2,3,4,5]))
print("======================")
print("The Mod is ::",st.mode([1,2,3,1,2,1,5]))
print("======================")
```

```python
import random as rd
print("The Choice num is ::",rd.choice([10,20,30,40,50]))
print("======================")
print("The Choice Text is ::",rd.choice(["Ahmed" , "Elsayed" ,"Ebrahim" , "Turkey"]))
print("======================")
print("The random num is ::",rd.random())
print("======================")
print("The randrange of (10) is ::",rd.randrange(10))
print("======================")
print("The randrenge of (10,20) is ::",rd.randrange(10,20))
print("======================")
print("The randrenge of (10,20,2) is ::",rd.randrange(10,20,2))
print("======================")
print("he randint num is ::",rd.randint(10,50))
print("=======================")
print("The sample of the range (0,10) of 2 ::",rd.sample(range(0,10),2))
print("======================")
print("The shuffle value is ::",rd.shuffle([1,2,3,4,5,6]))
print("======================")
print("The uniform value is ::",rd.uniform(1,10))
print("======================")
```
