```python
class Game :
    def __init__(self) :
        print("Welcome To Games App.......^__^")
        print("press [1] : To Play  Odd_Even Game ::")
        print("press [2] : To Play  Sum_Average Game ::")
        print("press [3] : To Play  Multiplication Game ::")
         
        self.Choice_Game()
        
    #############################################################################   
    
    def Choice_Game (self) :
         while True :
                
                
                User_Choice = input("Enter u Choice :")
                try :
                
                    User_Choice = int(User_Choice)
                    if User_Choice == 1 :
                        self.Even_Odd_Game()
                    elif User_Choice == 2 :
                        self.Sum_Average_Game()
                    elif User_Choice == 3 :
                        self.Multiplication_Game()
                    else :
                        print("Please Choice  1  ,  2  or 3 :")
                except ValueError :
                    print("Please Enter Valid Number :")
                    
    ################################################################################
    
    def Even_Odd_Game(self) :
        print("Wellcome to The Gime ::")
        print("Enter A Number And We will Tell U its Even or Odd Number :::")

        while True :
            
            Num = input("Enter The Number :")
            if Num == "X" :
                print("Closing The Game")
                print("Thnx")

                break

            try :
                Num = int (Num)

                if Num % 2 == 0 :
                    print("The Number Is Even")
                else :
                    print("The Numberb Is Odd")

            except ValueError :
                print("please Enter Valid Number:")

            print('========================================')

    ###################################################################
    
    def Sum_Average_Game(self):
        print("Wellcome To The Game ::")
        print("Enter How many Number?  and Enter Numbers ! And we Give U And Sum  Avarage ::")


        Count = int(input("How many Number U would To get Sum  and Avrage :"))

        Cur_count = 0
        Summ = 0

        while Cur_count < Count :
            Number = int(input("Enter The Number ::"))
            Cur_count +=1
            Summ +=Number
            print("=====================================")

        print("The Summ of Numbers = " , Summ)
        print("=====================================")
        print("The Average of Numbers = " , Summ/Count)
        
    ########################################################################
    
    def Multiplication_Game(self):
        print("Wellcome To Multitplication App::")
        print("Enter the start and end of multiblication App ::")
        Start  = int(input("Enter Start App::"))
        End    = int(input("Enter The End App::"))
        print("===============================")

        print("This Multiplication of ",Start)

        print("===============================")


        for X in range(Start,End+1) :
            for Y in range(1,13):
                print(X  , "*" , Y , "=" , X*Y)
            print("===============================")
            print("This Multiplication of  ",   X+1)
            print("===============================")


        print("Thnx To Us Multiplication App")
        
    
    
Game1=Game()
```
